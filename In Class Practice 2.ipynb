```python
from scipy import stats
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
print(stats.__doc__)
```

```python
mu, sigma = 100, 15


iq_scores = np.random.normal(
    loc=mu,
    scale=sigma,
    size=500)

count, bins, ignored = plt.hist(iq_scores, 30, normed=True)
plt.plot(
    bins, 
    1/(sigma * np.sqrt(2 * np.pi)) * np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
    linewidth=2, 
    color='r')

plt.show();
```

```python
import seaborn as sns
sns.set(style="whitegrid")
ax = sns.boxplot(x=iq_scores)
plt.show();
```
